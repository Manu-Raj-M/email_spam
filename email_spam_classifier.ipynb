```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mail_data.csv')
```

```python
data = df.where((pd.notnull(df['Category'])), '')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.loc[data['Category'] == 'spam', 'category'] = 0
data.loc[data['Category'] == 'ham', 'category'] = 1
```

```python
x = data['Message']
y = data['category'].astype('int')
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_Test = train_test_split(x, y, test_size=0.2, random_state=3)
# 80% train and 20% test
# If you set random_state=3, you’ll always get the same 80% and 20% split every time you run the code.
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_Test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
# stopwords are the words which do not add much meaning, i.e. we can safely ignore them while processing the text data
x_train_feature = feature_extraction.fit_transform(x_train)
x_test_feature = feature_extraction.transform(x_test)
```

```python
print(data['category'].astype('int'))
```

```python
y_train = y_train.astype('int')
y_Test = y_Test.astype('int')
```

```python
print(x_train_feature)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_feature, y_train)
```

```python
prediction_on_training_data = model.predict(x_train_feature)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
print(f"accuracy of the data: {accuracy_on_training_data}")
```

```python
predict_on_test_data = model.predict(x_test_feature)
accuracy_on_test_data = accuracy_score(y_Test, predict_on_test_data)
print(f"the accuracy on test data: {accuracy_on_test_data}")
```
